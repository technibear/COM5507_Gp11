```python
import pandas as pd
import string
import re
from nltk.tokenize import word_tokenize

encodings = ['utf-8', 'ISO-8859-1']
file_path = 'lyrics_data.csv'

for encoding in encodings:
    try:
        df = pd.read_csv(file_path, encoding=encoding)
        break
    except UnicodeDecodeError:
        continue

df['Lyrics'] = df['Lyrics'].fillna('')
df['Lyrics'] = df['Lyrics'].str.lower()
df['Lyrics'] = df['Lyrics'].str.replace('\[.*?\]', '', regex=True)  # 去除 [ ] 及其中的内容
df['Lyrics'] = df['Lyrics'].str.replace('[{}]'.format(string.punctuation), '', regex=True)  # 去除标点符号
df['Lyrics'] = df['Lyrics'].str.replace('\d+', '', regex=True)  # 去除数字
df['Lyrics'] = df['Lyrics'].str.strip()  # 去除前后的空格
df['tokens'] = df['Lyrics'].apply(word_tokenize)

df.to_csv('lyrics_data_cleaned.csv', mode='w', encoding='UTF-8', index=False)
```

```python
import csv
from nltk.sentiment import SentimentIntensityAnalyzer

# 创建情感分析器对象
sia = SentimentIntensityAnalyzer()

with open('lyrics_data_cleaned.csv', 'r', encoding='UTF-8') as input_file, \
        open('lyrics_data_sentiment.csv', 'w', encoding='UTF-8', newline='') as output_file:
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)

    # 读取原始文件的标题行
    original_header = next(reader)
    
    # 写入标题行到输出文件
    writer.writerow(original_header + ['positive', 'negative', 'neutral', 'compound'])
    
    # 逐行读取输入文件的数据
    for row in reader:
        # 获取文本内容
        text = row[1]
        
        # 使用VADER进行情感分析
        sentiment_scores = sia.polarity_scores(text)
        
        # 提取情感分析结果
        positive_score = sentiment_scores['pos']
        negative_score = sentiment_scores['neg']
        neutral_score = sentiment_scores['neu']
        compound_score = sentiment_scores['compound']
        
        # 将原始行数据和情感分析结果写入输出文件
        writer.writerow(row + [positive_score, negative_score, neutral_score, compound_score])
```
